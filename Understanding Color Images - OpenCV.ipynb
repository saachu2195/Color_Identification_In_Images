```python
## Importing the require libraries.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
## Importing the image from local disk.
img = cv2.imread("C:/Users/square/Desktop/Computer Vision/OpenCV/bruano.png")      

## Viewing the images.
plt.figure(figsize = (10, 8))
plt.imshow(img)
plt.show()
```

```python
## Converting the image into RGB.

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_rgb1 = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb1)
```

```python
## Convert the image into grayscale.

gray_img = cv2.cvtColor(img_rgb1, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap = 'gray')
```

```python
## Resizing the images.

resized_image = cv2.resize(img, (1200,600))
plt.imshow(resized_image)
```

```python
fig, axs = plt.subplots(nrows = 1 , ncols = 3, figsize = (20, 20))
for i in range(0, 3):
    ax = axs[i]
    ax.imshow(img_rgb1[:, :, i], cmap = 'gray')
plt.show()    
```

```python
## Converting the img into HSV & HSL.

img_hsl = cv2.cvtColor(img_rgb1, cv2.COLOR_RGB2HLS)
img_hsv = cv2.cvtColor(img_rgb1, cv2.COLOR_RGB2HSV)
fig, (ax1,ax2) = plt.subplots(nrows =1, ncols = 2, figsize = (20,20))
ax1.imshow(img_hsv)
ax2.imshow(img_hsl)
```

## DRAWING ON IMAGES


```python
## Reading the images.

wall_love = cv2.imread("C:/Users/square/Desktop/walloflove.jpg")
plt.imshow(wall_love)
```

```python
## Converting the image into RGB

wall_img = cv2.cvtColor(wall_love, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (12, 10))
plt.imshow(wall_img)
```

```python
## Defining a reactangle on images


wall_copy = wall_img.copy()
cv2.rectangle(wall_copy, pt1 = (80, 47), pt2 = (98, 53), color = (255, 0, 0), 
             thickness = 5)
plt.imshow(wall_copy)
```

```python
plt.figure(figsize = (12, 10))
plt.imshow(wall_copy)
```

```python
cv2.circle(wall_copy, center = (950, 50), radius = 50,
           color = (0, 0, 255), thickness = 5)
plt.imshow(wall_copy)
```

```python
cv2.putText(wall_copy, text = "the Wall of Love",
            org = (150, 250),
            fontFace = cv2.FONT_HERSHEY_DUPLEX,
            fontScale = 2,
            color = (0, 255, 0),
            thickness = 2,
            lineType = cv2.LINE_AA)
plt.imshow(wall_copy)
```
